```python
import matplotlib.pyplot as plt
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Scatter Plot Data: Work Experience vs Salary
work_experience = np.random.uniform(1, 30, 30)  # Years of experience
salary = work_experience * np.random.uniform(2000, 5000, 30) + np.random.normal(0, 10000, 30)  # Salary with noise

# Bubble Plot Data: R&D Investment vs Number of Patents
investment = np.random.randint(5, 100, 30)  # R&D investment in million dollars
patents = investment * np.random.uniform(0.5, 2.0, 30)  # Number of patents
bubble_sizes = patents * 10  # Bubble size based on patents

# Create figure and axes
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Scatter Plot: Work Experience vs Salary
axes[0].scatter(work_experience, salary, c='blue', alpha=0.7, edgecolors='black')
axes[0].set_title("Scatter plot:Work Experience vs Salary", fontsize=14, fontweight='bold', color='blue')
axes[0].set_xlabel("Years of Work Experience", fontsize=12)
axes[0].set_ylabel("Salary ($)", fontsize=12)
axes[0].grid(True, linestyle='--', alpha=0.6)

# Bubble Plot: R&D Investment vs Patents
scatter = axes[1].scatter(investment, patents, s=bubble_sizes, c=investment, cmap='viridis', alpha=0.7, edgecolors='black')
axes[1].set_title("Bubble plot:R&D Investment vs Patents", fontsize=14, fontweight='bold', color='green')
axes[1].set_xlabel("R&D Investment (Million $)", fontsize=12)
axes[1].set_ylabel("Number of Patents Filed", fontsize=12)
axes[1].grid(True, linestyle='--', alpha=0.6)

# Add colorbar for bubble plot
cbar = plt.colorbar(scatter, ax=axes[1])
cbar.set_label("Investment Level ($M)", fontsize=12)

# Show plots
plt.tight_layout()
plt.savefig('plot2.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data: Monthly revenue (in units, no currency symbol)
months = np.arange(1, 13)  # 12 months

product_sales = np.random.randint(20, 50, 12)  # Product sales revenue (units)
subscriptions = np.random.randint(10, 30, 12)  # Subscription-based revenue (units)
advertising = np.random.randint(5, 20, 12)  # Ad revenue (units)

# Create stacked bar plot
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(months, product_sales, label='Product Sales', color='pink')
ax.bar(months, subscriptions, bottom=product_sales, label='Subscriptions', color='purple')
ax.bar(months, advertising, bottom=product_sales + subscriptions, label='Advertising', color='green')

# Formatting
ax.set_title("Monthly Revenue Breakdown", fontsize=14, fontweight='bold')
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Revenue (Units)", fontsize=12)
ax.set_xticks(months)
ax.legend(loc="upper left")
ax.grid(axis='y', linestyle='--', alpha=0.6)
plt.savefig('plot5.png', dpi=300, bbox_inches='tight')
# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data: Random sample for the variables
np.random.seed(42)
x = np.random.normal(50, 15, 100)  # Variable 1
y = np.random.normal(40, 10, 100)  # Variable 2

# Create the figure and axis
fig = plt.figure(figsize=(6, 6))

# Main scatter plot
ax = fig.add_subplot(111)

# Scatter plot
ax.scatter(x, y, alpha=0.7, color='blue')

# Marginal histograms
# Top histogram for variable X
ax_histx = fig.add_axes([0.1, 0.85, 0.8, 0.1])  # (left, bottom, width, height)
ax_histx.hist(x, bins=20, color='blue', alpha=0.7)
ax_histx.set_xlim(ax.get_xlim())

# Right histogram for variable Y
ax_histy = fig.add_axes([0.85, 0.1, 0.1, 0.8])  # (left, bottom, width, height)
ax_histy.hist(y, bins=20, color='blue', alpha=0.7, orientation='horizontal')
ax_histy.set_ylim(ax.get_ylim())

# Formatting
ax.set_title("Marginal Plot with Scatter and Histograms", fontsize=14, fontweight='bold')
ax.set_xlabel("Variable X", fontsize=12)
ax.set_ylabel("Variable Y", fontsize=12)

# Adjust layout manually to avoid overlapping
plt.subplots_adjust(left=0.1, right=0.9, top=0.85, bottom=0.15)
plt.savefig('plot10.png', dpi=300, bbox_inches='tight')
# Show plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

# Load Iris dataset
iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)

# Add target column (species)
iris['species'] = iris_data.target
iris['species'] = iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Create a pairplot with KDE on the diagonal, different markers for different species, and a new color palette
sns.pairplot(iris, hue="species", diag_kind="kde", markers=["o", "s", "D"], palette="husl", plot_kws={'alpha': 0.7})

# Set the title with a new message and adjusted position
plt.suptitle("Pairwise Relationships in Iris Dataset with KDE", y=1.05)

# Display grid for better readability
plt.subplots_adjust(top=0.9)  # Adjust top margin to make room for the title
plt.savefig('plot14.png', dpi=300, bbox_inches='tight')
# Show the plot
plt.show()
```
